### Forks, Chains, Colliders

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Independency

일반적으로 X에 대한 지식이 Y에 대한 지식을 바꾸지 않을 때 (또는 그 반대의 경우도 마찬가지) 두 변수, 즉 X와 Y는 독립적이라고 말한다.

$$
\begin{align}
P(Y) = P(Y|X) \\
P(X) = P(X|Y)
\end{align}
$$

두 변수 간의 독립에 대한 한 가지 흥미로운 결과는 두 변수의 결합 분포가 주변확률의 곱으로 인수 분해된다는 것이다.

$$P(X,Y) = P(X)P(Y)$$

$⫫$를 사용하면 $X$와 $Y$가 독립적이라는 사실은 다음과 같이 표현할 수 있다.

$$X ⫫ Y$$

독립성의 개념은 통계와 인과관계에서 중요한 역할을 한다. 이 개념이 일반화된 조건부 독립(conditional independence)은 훨씬 더 중요하다. $Z$를 관찰했다고 가정했을 때, $X$가 $Y$에 대한 새로운 정보를 제공하지 않는다면 우리는 $Z$가 주어졌을 때 $X$와 $Y$가 조건부 독립이라고 한다.

$$X ⫫ Y|Z$$

같은 의미로 확률로 표현하면 다음과 같다.

$$P(X,Y|Z) = P(X|Z)P(Y|Z)$$

확률 분포에서 독립이라는 것을 나타내기 위해 다음과 같은 표현을 사용한다.

$$X ⫫_P Y$$

반면에 그래프에서 $X$와 $Y$가 독립이라는 것을 나타내기 위해 다음의 표현을 사용한다.

$$X ⫫_G Y$$

### 인과관계 추론의 조건

인과추론의 관점에서 볼 때, 그래프의 조건부 독립을 통계적 독립으로 매핑할 수 있는지를 확인해야 한다. 이를 위해서는 Causal Markov Condition을 충족해야 한다. Causal Markoc Condition은 node $V_i$의 부모가 주어졌을 때 $V_i$의 부모를 제외한 모든 비 자손 노드와 독립이다. 따라서 수식적으로 다음과 같이 표현할 수 있다.

$$V_i ⫫_G V_j \Vert PA(V_i) \forall_{j \neq i \in G(V,E) \{DE(V_i), PA(V_i)\}}$$

중요한 것은 인과적 마르코프 조건이 성립할 때 다음이 참이 된다

$$X ⫫_G Y|Z \Rightarrow X ⫫_P Y|Z$$

이를 전역 마르코프 성질(Global Markov Property)라고 한다.

### Chain

```python
NOISE_LEVEL = 0.2
N_SAMPLES = 1000
```

```python
a = np.random.randn(N_SAMPLES)
b = a + NOISE_LEVEL * np.random.randn(N_SAMPLES)
c = b + NOISE_LEVEL * np.random.randn(N_SAMPLES)
```

```python
fig, ax = plt.subplots(figsize = (8, 5))
ax.scatter(a, b, marker = '.', edgecolor = 'none')
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(False)
ax.yaxis.grid(True, ls = ':', alpha = 0.5) 
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))
ax.scatter(b, c, marker = '.', edgecolor = 'none')
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(False)
ax.yaxis.grid(True, ls = ':', alpha = 0.5) 
plt.xlabel('b')
plt.ylabel('c')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))
ax.scatter(a, c, marker = '.', edgecolor = 'none')
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(False)
ax.yaxis.grid(True, ls = ':', alpha = 0.5) 
plt.xlabel('a')
plt.ylabel('c')
plt.show()
```
